```python
#Countries USA, Singapore, UK, Isreal, India, China, Russia, Germany, France
locations = ["USA" , "Singapore", "UK", "Israel", "India", "China", "Russia", "Germany", "France"]
geolocation = ["38.50, -98.0,500km", "1.290270, 103.851959, 20km", "51.509865,  -0.118092, 200km", "31.771959, 35.217018, 80km",\
              "28.644800, 77.216721, 500km", "39.916668, 116.383331, 500km", "55.751244, 37.618423, 500km", "52.520008, 13.404954, 500km",\
              "48.864716, 2.349014, 500km"]
Company = ["tesla or TSLA", "bmw OR BMW", "volkswagen OR Volkswagen"]
```

```python
count = 0
for x in geolocation:
    for y in Company:
        t = twint.Config()
        t.Verified = True #Verified vs unverified
        t.Search = y
        t.Pandas = True
        t.Since = "2022-01-19"
        t.Geo=x
        t.Lang = 'en'
        t.Profile_full = True
        t.Store_csv = True
        twint.run.Search(t)
        filename = y +'_' + locations[count]  + '.csv'
        twint.storage.panda.Tweets_df.to_csv(filename)
    count = count +1
```

```python
import pandas as pd
import glob
import os

# merging the files
joined_files = os.path.join("C:/Users/yodar/Desktop/Info retrieval/New", "*.csv")
  
# A list of all joined files is returned
joined_list = glob.glob(joined_files)
  
# Finally, the files are joined
df = pd.concat(map(pd.read_csv, joined_list), ignore_index=True)
df.head()
```

```python
df.columns
```

```python
df = df.loc[:, ["id", 'username','geo', 'created_at', 'date', 'tweet', 'language', 'hashtags', 'cashtags', 'link', 'urls', 'photos', 'video', 'nlikes', 'nreplies', 'nretweets']]
```

```python
df["date"] = pd.to_datetime(df["date"])
print(df.dtypes)
```

```python
df = df.sort_values(by='date',ascending = False)
```

```python
df1 = df.drop_duplicates(subset=['tweet'], keep='first')
```

```python
df1 = df1.loc[df1['language'] == 'en']
```

```python
df1 = df1.reset_index()
```

```python
# df1.drop('index', axis=1, inplace=True)
```

```python
# Remove the whitespace in geolocations 
geolocation = [geolocation[i].replace(" ", "") for i in range(len(geolocation))]
geo_location_map = {geolocation[i]: locations[i] for i in range(len(locations))}

data = df1
data['geo'] = np.vectorize(lambda x: geo_location_map[x])(data['geo'])
```
