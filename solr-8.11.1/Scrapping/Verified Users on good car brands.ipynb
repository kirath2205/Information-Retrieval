```python
companies = ["Audi", "Bentley", "Cadillac", "Ferrai", "Lambo", "Maserati", "Mercedes", "Porsche", "RollsRoyce", "Volkswagen", "Volvo"]
```

```python
for x in companies:
    t = twint.Config()
    t.Verified = True
    t.Search = x
    t.Pandas = True
    t.Since = "2022-03-25" #past 10 days
    t.Lang = 'en'
    t.Profile_full = True
    t.Store_csv = True
    twint.run.Search(t)
    name = x + '.csv'
    twint.storage.panda.Tweets_df.to_csv(name)
```
